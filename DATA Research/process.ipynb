```python
import numpy as np
import plotly.graph_objects as go
import os

# Ensure the directory to store images exists
output_dir = "pendulum_frames_"
os.makedirs(output_dir, exist_ok=True)

# Constants
g = 9.81
L = 2
mu = 0.1
theta_0 = np.pi / 3
theta_dot_0 = 0

def get_theta_double_dot(theta, theta_dot):
    return -mu * theta_dot - (g / L) * np.sin(theta)

def simulate_pendulum_motion(t_max, delta_t):
    theta = theta_0
    theta_dot = theta_dot_0
    theta_values = []
    theta_dot_values = []
    
    for time in np.arange(0, t_max, delta_t):
        theta_values.append(theta)
        theta_dot_values.append(theta_dot)
        
        theta_double_dot = get_theta_double_dot(theta, theta_dot)
        theta += theta_dot * delta_t
        theta_dot += theta_double_dot * delta_t

    return np.array(theta_values), np.array(theta_dot_values)

# Parameters
t_max = 10
delta_t_values = np.arange(0.1, 0.001, -0.001)

# Create frames and save each as an image
frame_count = 0
for delta_t in delta_t_values:
    theta_values, theta_dot_values = simulate_pendulum_motion(t_max, delta_t)
    
    fig = go.Figure(
        data=[go.Scatter(x=theta_values, y=theta_dot_values, mode='lines', line=dict(color='darkorange'))],
        layout=go.Layout(
            title=f"Pendulum - Time Step (delta_t): {delta_t:.4f}",
            xaxis_title="Angle θ (radians)",
            yaxis_title="Angular Velocity θ̇ (radians/s)",
            width=700,
            height=700,
            xaxis=dict(scaleanchor="y"),
            yaxis=dict(constrain='domain')
        )
    )

    # Save each frame as a PNG
    frame_file = os.path.join(output_dir, f"frame_{frame_count:04d}.png")
    print("*")
    fig.write_image(frame_file)
    frame_count += 1
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Constants
g = 9.81  # gravitational acceleration (m/s^2)
L = 2     # length of pendulum (m)
mu = 0.1  # damping coefficient

# Initial conditions
theta_0 = np.pi / 3  # initial angle (60 degrees)
theta_dot_0 = 0      # initial angular velocity

# Function to calculate angular acceleration
def get_theta_double_dot(theta, theta_dot):
    return -mu * theta_dot - (g / L) * np.sin(theta)

# Function to simulate the pendulum motion and record θ and θ̇
def simulate_pendulum_motion(t_max, delta_t):
    theta = theta_0
    theta_dot = theta_dot_0
    theta_values = []
    theta_dot_values = []
    
    for time in np.arange(0, t_max, delta_t):
        theta_values.append(theta)
        theta_dot_values.append(theta_dot)
        
        theta_double_dot = get_theta_double_dot(theta, theta_dot)
        theta += theta_dot * delta_t
        theta_dot += theta_double_dot * delta_t

    return np.array(theta_values), np.array(theta_dot_values)

# Directory for output frames
output_dir = "pendulum_frames_centered"
os.makedirs(output_dir, exist_ok=True)

# Simulation parameters
t_max = 30
delta_t_values = np.arange(1.0, 0.001, -0.001)  # Decremental range

# Comments at different stages of the simulation
comments = [
    "Initial oscillations are broad, indicating high energy.",
    "Damping effects gradually reduce oscillation amplitude.",
    "As time progresses, pendulum motion stabilizes and slows.",
    "Approaching the end: the pendulum reaches near-equilibrium, with minimal motion."
]

# Generate and save each frame with dynamically centered plot
for i, delta_t in enumerate(delta_t_values):
    theta_values, theta_dot_values = simulate_pendulum_motion(t_max, delta_t)
    
    # Calculate dynamic axis limits based on data range with some padding
    x_limit = max(abs(theta_values)) * 1.1
    y_limit = max(abs(theta_dot_values)) * 1.1
    
    # Plot with a large figure size
    plt.figure(figsize=(24, 24), dpi=80)  # Large figure size for detailed plot
    plt.plot(theta_values, theta_dot_values, color='darkorange', linewidth=5)
    
    # Set plot title, axis labels, and limits
    plt.title(f"Damped Pendulum Simulation - Phase Space Trajectory\nCreated by Mayank Dwivedi(Data Scientist)", fontsize=22, pad=20)
    plt.suptitle(f"Simulating a damped pendulum's angular displacement and velocity over time using Simple ODE modeling.\n delta_time : {delta_t} secs", fontsize=20, y=0.95)
    plt.xlabel("Angle θ (radians)", fontsize=18)
    plt.ylabel("Angular Velocity θ̇ (radians/s)", fontsize=18)
    
    # Set dynamic axis limits to fit the plot at the center of the figure
    plt.xlim(-x_limit, x_limit)
    plt.ylim(-y_limit, y_limit)
    
    # Determine which comment to display based on frame position
    comment_index = min(i // (len(delta_t_values) // len(comments)), len(comments) - 1)
    plt.figtext(0.5, 0.02, comments[comment_index], wrap=True, horizontalalignment='center', fontsize=20, color='green')
    
    # Save the frame as an image
    frame_file = os.path.join(output_dir, f"frame_{i:04d}.png")
    plt.savefig(frame_file, dpi=80)
    plt.close()  # Close plot to save memory
    import gc
    print(gc.collect(), end="\r")
print("Frames saved successfully.")

```
